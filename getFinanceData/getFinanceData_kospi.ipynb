```python
def preprocess(df: pd.DataFrame, tic: str)-> pd.DataFrame:
    df.reset_index(inplace=True)
    df['tic'] = tic
    df.rename(columns={'Date': 'datadate'}, inplace=True)
    df['datadate'] = df['datadate'].apply(lambda x: str(x))
    df.datadate.replace('-','',regex=True, inplace=True)
    df.datadate.replace('00:00:00','',regex=True, inplace=True)
    df.rename(columns={'Open': 'prcod'}, inplace=True)
    df.rename(columns={'Close': 'prccd'}, inplace=True)
    df.rename(columns={'High': 'prchd'}, inplace=True)
    df.rename(columns={'Low': 'prcld'}, inplace=True)
    df.rename(columns={'Volume': 'cshtrd'}, inplace=True)
    df['ajexdi'] = 1
    df = df[['datadate', 'tic', 'prccd','ajexdi', 'prcod', 'prchd', 'prcld', 'cshtrd']]
    return df
```
