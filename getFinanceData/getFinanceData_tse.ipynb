```python
def preprocess(df: pd.DataFrame, tic: str)-> pd.DataFrame:
    df.reset_index(inplace=True)
    df['tic'] = tic
    df.rename(columns={'Date': 'datadate'}, inplace=True)
    df['datadate'] = df['datadate'].apply(lambda x: str(x))
    df.datadate.replace('-','',regex=True, inplace=True)
    df.datadate.replace('00:00:00','',regex=True, inplace=True)
    df.rename(columns={'Open': 'prcod'}, inplace=True)
    df.rename(columns={'Close': 'prccd'}, inplace=True)
    df.rename(columns={'High': 'prchd'}, inplace=True)
    df.rename(columns={'Low': 'prcld'}, inplace=True)
    df.rename(columns={'Volume': 'cshtrd'}, inplace=True)
    df['ajexdi'] = 1
    df = df[['datadate', 'tic', 'prccd','ajexdi', 'prcod', 'prchd', 'prcld', 'cshtrd']]
    return df
```

```python
import collections
stocks = collections.defaultdict()
cnt = 0
dayNumber = len(pdr.get_data_yahoo('7203.T', '2009-01-01', '2020-09-30'))
for i in tqdm(range(36)):
    #tmp = fdr.DataReader(tse_stocks.Symbol[i], '2009-01-01', '2020-08-31',exchange='TSE')
    #dayNumber = len(fdr.DataReader('7203', '2009-01-01', '2020-08-31',exchange='TSE'))
    tmp = pdr.get_data_yahoo(tse_stocks.Symbol[i]+'.T', '2009-01-01', '2020-09-30')
    if len(tmp)==dayNumber:
        stocks[i+1] = preprocess(tmp, tse_stocks.Symbol[i])
    else:
        stocks[i+1] = pd.DataFrame()
        cnt += 1
    time.sleep(0.5)
print(cnt)
```

```python
tse30 = pd.DataFrame()
tse30Index = 0
cnt = 0
for i in range(1, len(stocks)+1):
    tse30 = pd.concat([tse30,stocks[i]])
    if stocks[i].empty == False:
        cnt += 1
        tse30Index = tse30Index + stocks[i].prccd

tse30Index = tse30Index/30
tse30.reset_index(drop=True,inplace=True)
tse30.to_csv('/home/kongminseok/hdd/research/DRLFAST/data/tse_30_2009_2020.csv',encoding='utf-8-sig')
display(tse30)
```
